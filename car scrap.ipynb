```python
data=data.replace('-',np.nan)
data=data.dropna()
data[data['name'] == '-']
```

```python
data.shape
```

```python
#extracting year from name 
name1=[]
for i in data['name']:
    name1.append(i.split())
    
y=[sublist[0:1] for sublist in name1]

#flattening the list(merge sublists into main list 
year=[]
for i in y:
    for j in i:
        year.append(j)

#creating manufacture company column       
manfctr=[sublist[1:2] for sublist in name1]

#flattening the list(merge sublists into main list )
manufacture=[]
for i in manfctr:
    for j in i:
        manufacture.append(j)
        
#removing km from kilometer driven       
km=[]
for i in data['kilometers_driven']:
    km.append(i.split())
    
#price    
price=[]

for i in data['price']:
    price.append(i.split('₹'))

#price
data['price']=[sublist[1] for sublist in price]
    
#km
data['kilometers_driven']=[sublist[0] for sublist in km]

#manufactured by
data['manufacture']=manufacture

#year column
data['year']=year
```
